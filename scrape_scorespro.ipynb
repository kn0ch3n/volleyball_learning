```python
import lxml.html
from cssselect import HTMLTranslator, SelectorError
from lxml import etree # debugging

def get_results(url):
    """Extracts results from an url."""
    doc = get_url(url)
    #print(etree.tostring(doc))
    tables = get_result_tables(doc)
    results = [extract_results(table) for table in tables]
    results = filter(lambda x: x != {}, results) # remove empty dicts
    return results

def get_url(url):
    """Returns the HTML at url as lxml.etree._ElementTree."""
    doc = lxml.html.parse(url)
    return doc

def cssselector(pattern):
    """Takes a CSS pattern, and returns it converted to XPath.
    The resulting pattern can be used in lxml.etree._ElementTree.xpath()."""
    try:
        expression = HTMLTranslator().css_to_xpath(pattern)
    except SelectorError:
        print('ERROR: Invalid selector.')
    return expression

def get_result_tables(doc):
    """Returns a list of all tables in doc matching cssselector("div.compgrp > table.gteam")."""
    table_pattern = cssselector("div.compgrp > table.gteam")
    tables = [table for table in doc.xpath(table_pattern)]
    #print(len(tables))
    return tables

def extract_results(table):
    """Extracts game data from a result table.
    Returns the data as a dict."""
    #print(etree.tostring(table))
    trs = table.xpath(cssselector("tr"))
    home_team = trs[0].xpath(cssselector("td.hometeam"))[0].text_content().strip()
    home_score = trs[0].xpath(cssselector("td.ts_setB"))[0].text_content().strip()
    away_team = trs[1].xpath(cssselector("td.awayteam"))[0].text_content().strip()
    away_score = trs[1].xpath(cssselector("td.ts_setB"))[0].text_content().strip()
    try:
        d = {"team1": home_team, "team2": away_team,
         "score1": int(home_score), "score2": int(away_score),
         "winner": home_team if home_score > away_score else away_team}
        return d
    except:
        return {}
```

#Women World Championship
Get 2010 and 2014 results from http://www.scorespro.com/.
Firstly, construct the urls.

```python
import sys
prefixes = ["http://www.scorespro.com/volleyball/world/world-championship-women"]
years = ["2010", "2014"]
# TODO: get rounds from dropdown menu on the site
rounds = ["first-round", "second-round", "final-round",
          "african-qualification", "asian-and-oceania-qualification",
          "european-qualification-r1", "european-qualification-r2",
          "european-qualification-r3", "norceca-qualification-r3",
          "norceca-qualification", "south-american-qualification",
          "norceca-qualification-r2", "norceca-qualification-r2-play-offs"]

# simple hack to avoid getting dates:
# use only the year instead
results = dict()
for year in years:
    urls = ["/".join([prefix, year, round, ""])
        for prefix in prefixes for round in rounds for suffix in suffixes]
    results[year] = []
    for url in urls:
        result = get_results(url)
        if len(result) > 0:
            results[year].extend(result)
```
